```python
import yfinance as yf
import pandas as pd
```

```python
etfs = ['BILS', 'USFR', 'BIL', 'SGOV', 'IEF', 'TLT']
start_date = '2021-01-01'
end_date = '2024-07-25'
```

```python
def get_historical_prices(etfs, start_date, end_date):
    price_data = {}
    for etf in etfs:
        ticker = yf.Ticker(etf)
        hist = ticker.history(start=start_date, end=end_date)
        hist = hist.reset_index()
        hist['Date'] = hist['Date'].dt.date
        price_data[etf] = hist.set_index('Date')['Close']
    # Combine all the price data into a single DataFrame
    combined_data = pd.DataFrame(price_data)
    return combined_data

```

```python
historical_prices = get_historical_prices(etfs, start_date, end_date)
historical_prices.to_csv('/Users/liyihan/Desktop/Work/ASL/ETF Hedging Data/ETF/processed data/adjNAV.csv')
```

```python
def get_dividend_data(etfs, start_date, end_date):
    dividend_data = {}
    for etf in etfs:
        ticker = yf.Ticker(etf)
        dividends = ticker.dividends
        # Filter the dividends by date range
        filtered_dividends = dividends[(dividends.index >= start_date) & (dividends.index <= end_date)]
        filtered_dividends.index = filtered_dividends.index.date  # Convert to date format
        dividend_data[etf] = filtered_dividends
    # Combine all the dividend data into a single DataFrame
    combined_dividends = pd.DataFrame(dividend_data)
    return combined_dividends
```

```python
dividend_data = get_dividend_data(etfs, start_date, end_date)
dividend_data.to_csv('/Users/liyihan/Desktop/Work/ASL/ETF Hedging Data/ETF/processed data/etf_dividends.csv', index=True)
```
